# Import libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Reading the Data from file

```python
df = pd.read_csv('Data/data_science_salary_2021.csv')
```

## Inspect the data

```python
df.info()
df.head()
```

What is the mean 'Avg salary(K)' according to "job_title_sim" ?

```python
ax = sns.barplot(data=df, x='job_title_sim', y='Avg Salary(K)')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha = 'right');
plt.savefig('barplot.png', pad_inches = 1, bbox_inches = 'tight')
#plt.tick_params(labelrotation = 45)
```

Who is using Python most according to "job_title_sim" ?

```python
ax = sns.barplot(data=df, x='job_title_sim', y='Python')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha = 'right');
plt.savefig('barplot2.png', pad_inches = 1, bbox_inches = 'tight')
#plt.tick_params(labelrotation = 45)
```
